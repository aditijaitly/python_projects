```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
from matplotlib import pyplot as plt 

cars = pd.read_csv('cars.csv')
cars.head(5)
cars.tail(5)
```

```python
cars.dtypes
```

```python
#removing unecessary columns
cars = cars.drop(['Engine Fuel Type','Market Category','Vehicle Style','Popularity','Number of Doors','Vehicle Size'],axis = 1)
cars.head(5)

```

```python
cars.dtypes
```

```python
#rename columns
cars = cars.rename(columns = {"Engine HP":"HP","Engine Cylinders":"Cyclinders","Transmission Type":"Transmission","Driven_Wheels":"Drive Mode","highway MPG":"MPG-H","city mpg":"MPG-C","MSRP":"Price"})
cars.head()
```

```python
cars.shape
```

```python
duplicate_rows = cars[cars.duplicated()]
print("Number of duplicate rows :",duplicate_rows.shape)
```

```python
cars.count()
```

```python
cars = cars.drop_duplicates()
```

```python
cars.count()
```

```python
print(cars.isnull().sum())
```

```python
cars = cars.dropna()
```

```python
cars.count()
```

```python
print(cars.isnull().sum())
```

```python
# finding IQR(Inter quartile range to remove outliers)
numeric_columns = cars.select_dtypes(include=['number'])

Q1= numeric_columns.quantile(0.25)
Q3= numeric_columns.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
cars = cars[~((numeric_columns < (Q1 - 1.5 * IQR)) | (numeric_columns > (Q3 + 1.5 * IQR))).any(axis=1)] 
cars.shape
```

```python
cars.Make.value_counts().nlargest(40).plot(kind ='bar',figsize=(10,5))
plt.title("Number of cars by make")
plt.ylabel('Number of cars')
plt.xlabel('Make')
```

```python
numeric_columns = cars.select_dtypes(include=['number'])
plt.figure(figsize=(10,5))
c = numeric_columns.corr()
sns.heatmap(c,cmap="BrBG",annot=True)
c
```
